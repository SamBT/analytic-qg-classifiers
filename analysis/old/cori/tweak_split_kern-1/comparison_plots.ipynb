```python
import ROOT
```

```python
fq_soft = ROOT.TFile("/global/project/projectdirs/atlas/sambt/pythia83-output/optimal-classifiers/kernel-1/H2qq-CFCA3-split-onlysoft/total.root")
fq_coll = ROOT.TFile("/global/project/projectdirs/atlas/sambt/pythia83-output/optimal-classifiers/kernel-1/H2qq-CFCA3-split-onlycoll/total.root")
fq_fin = ROOT.TFile("/global/project/projectdirs/atlas/sambt/pythia83-output/optimal-classifiers/kernel-1/H2qq-CFCA3-split-onlyfin/total.root")

fg_soft = ROOT.TFile("/global/project/projectdirs/atlas/sambt/pythia83-output/optimal-classifiers/kernel-1/H2gg-CFCA3-split-onlysoft/total.root")
fg_coll = ROOT.TFile("/global/project/projectdirs/atlas/sambt/pythia83-output/optimal-classifiers/kernel-1/H2gg-CFCA3-split-onlycoll/total.root")
fg_fin = ROOT.TFile("/global/project/projectdirs/atlas/sambt/pythia83-output/optimal-classifiers/kernel-1/H2gg-CFCA3-split-onlyfin/total.root")

fq_reg = ROOT.TFile("/global/project/projectdirs/atlas/sambt/pythia83-output/optimal-classifiers/kernel-1/H2qq-CF3.0CA3.0/total.root")
fg_reg = ROOT.TFile("/global/project/projectdirs/atlas/sambt/pythia83-output/optimal-classifiers/kernel-1/H2gg-CF3.0CA3.0/total.root")
```

```python
tq_soft = fq_soft.Get("EventTree")
tq_coll = fq_coll.Get("EventTree")
tq_fin = fq_fin.Get("EventTree")

tg_soft = fg_soft.Get("EventTree")
tg_coll = fg_coll.Get("EventTree")
tg_fin = fg_fin.Get("EventTree")

tq_reg = fq_reg.Get("EventTree")
tg_reg = fg_reg.Get("EventTree")

tq_soft.SetLineColor(1)
tq_coll.SetLineColor(1)
tq_fin.SetLineColor(1)
tq_reg.SetLineColor(1)

tg_soft.SetLineColor(3)
tg_coll.SetLineColor(3)
tg_fin.SetLineColor(3)
tg_reg.SetLineColor(3)
```

```python
ROOT.gStyle.SetPalette(55)
c1 = ROOT.TCanvas("c1","c1")
c1.Divide(2,2)
c1.cd(1)
tq_soft.Draw("pjet_pt[0]")
tg_soft.Draw("pjet_pt[0]","","same")
c1.cd(2)
tq_fin.Draw("pjet_pt[0]","","")
tg_fin.Draw("pjet_pt[0]","","same")
c1.cd(3)
tq_coll.Draw("pjet_pt[0]","","")
tg_coll.Draw("pjet_pt[0]","","same")
c1.cd(4)
tq_reg.Draw("pjet_pt[0]","","")
tg_reg.Draw("pjet_pt[0]","","same")
c1.Draw()
c1.SaveAs("compare_pt_vary_splitting.png")
```

```python
ROOT.gStyle.SetPalette(55)
c2 = ROOT.TCanvas("c2","c2")
c2.Divide(2,2)
c2.cd(1)
tq_soft.Draw("pjet_eta[0]")
tg_soft.Draw("pjet_eta[0]","","PLC same")
c2.cd(2)
tq_fin.Draw("pjet_eta[0]","","")
tg_fin.Draw("pjet_eta[0]","","PLC same")
c2.cd(3)
tq_coll.Draw("pjet_eta[0]","","")
tg_coll.Draw("pjet_eta[0]","","PLC same")
c2.cd(4)
tq_reg.Draw("pjet_eta[0]","","")
tg_reg.Draw("pjet_eta[0]","","PLC same")
c2.Draw()
```
