```python
from tqdm import tqdm
import sys
sys.path.append("/home/sambt/analytic-qg-classifiers/analysis/utils/")
from master import *
prefix = '/home/sambt/pythia83-samples/optimal-classifiers/kernel-1/'
```

```python
#scanning over variations of c0/z*exp(c1z + c2z^2) for CF = 1.3
CF = 4/3
CA = 3.0
c0s = [1,5]
c1s = [1,10]
c2s = [1,10]
all_rocs = {}
all_aucs = {}
all_labels = {}
for c0 in tqdm(c0s):
    for c1 in c1s:
        for c2 in c2s:
            print(c0,c1,c2)
            label = "sq0{0}00-esq00{1}{2}".format(c0,c1,c2)
            f_opt = lambda x : c0*np.log(CF/CA)*np.count_nonzero(x,axis=1) + np.sum(c2*np.power(x,2,where=x>0,out=np.zeros(x.shape)) + c1*x,axis=1)
            rocs,aucs,labels = analysis.run_pipeline('H2qq-CF1.3CA3.0-'+label+'-r0.4','H2gg-CF3.0CA3.0-sqg0100-r0.4',f_opt)
            all_rocs[label] = rocs
            all_aucs[label] = aucs
            all_labels[label] = labels
```

```python
fig = plt.subplots(4,2,figsize=(40,20))
ct = 1
for c0 in tqdm(c0s):
    for c1 in c1s:
        for c2 in c2s:
            label = "sq0{0}00-esq00{1}{2}".format(c0,c1,c2)
            c0_str = str(c0) if c0 > 1 else "1"
            c1_str = str(c1) if c1 > 1 else ""
            c2_str = str(c2) if c2 > 1 else ""
            split = r'$P_{q \to qg} \propto \frac{'+c0_str+'}{z}e^{'+c1_str+'z + '+c2_str+'z^2}$'
            print(420+ct)
            ax = plt.subplot(240+ct)
            rocs,aucs,labels = all_rocs[label],all_aucs[label],all_labels[label]
            plotting.plot_many(rocs,aucs,labels,ax)
            plotting.stamp(0.02,0.6,line_3=r'$\sqrt{s} = 250$ GeV, $C_F = \frac{4}{3}$, $C_A = 3$')
            plt.text(0.02,0.35,split,fontsize=24)
            ct+=1
```
