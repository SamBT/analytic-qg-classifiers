```python
#Scan with R = 0.4
ez2s = [1,5,10,100]
roc_ez2s = {}
auc_ez2s = {}
lab_ez2s = {}
for ez2 in ez2s:
    f_opt = lambda x : np.sum(ez2*np.power(x,2,where=x>0,out=np.zeros(x.shape)) + x,axis=1)
    rocs,aucs,labels = analysis.run_pipeline('H2qq-CF3.0CA3.0-sqg0100-esq001{0}-r0.4'.format(ez2),'H2gg-CF3.0CA3.0-sqg0100-r0.4',f_opt)
    roc_ez2s[ez2] = rocs
    auc_ez2s[ez2] = aucs
    lab_ez2s[ez2] = labels
```

```python
importlib.reload(plotting)
fig, axes = plt.subplots(2,2,figsize=(20,20))
for i,ez2 in enumerate(ez2s):
    ax = plt.subplot(220+i+1)
    rocs,aucs,labels = roc_ez2s[ez2],auc_ez2s[ez2],lab_ez2s[ez2]
    plotting.plot_many(rocs,aucs,labels,ax)
    plotting.stamp(0.02,0.6)
    split = r'$P_{q \to qg} \propto e^{z+'+str(ez2)+'z^2}$' if ez2 > 1 else r'$P_{q \to qg} \propto e^{z+z^2}$'
    plt.text(0.02,0.35,split,fontsize=24)
plt.savefig("plots/cz2_scan_esq001n_r0p4.pdf")
```
