```python
import sys
from tqdm import tqdm
sys.path.append("/home/sambt/analytic-qg-classifiers/analysis/utils/")
from master import *
prefix = '/home/sambt/pythia83-samples/optimal-classifiers/kernel-1/'
```

```python
h2qq_samples = {"qgsame":prefix+'H2qq-CF3.0CA3.0-sqg0100/total.root', 
                "esq0001":prefix+'H2qq-CF3.0CA3.0-sqg0100-esq0001/total.root', 
                "esq0011":prefix+'H2qq-CF3.0CA3.0-sqg0100-esq0011/total.root', 
                "esq1000":prefix+'H2qq-CF3.0CA3.0-sqg0100-esq1000/total.root'}
fg = prefix+'H2gg-CF3.0CA3.0-sqg0100/total.root'
samples = list(h2qq_samples.keys())
```

```python
z_collection = {}
eta_collection = {}
phi_collection = {}
dR_collection = {}
label_collection = {}
for samp, fq in h2qq_samples.items():
    X,Y = tools.pre_process(fq,fg,nev_max=50000)
    z = X[:,:,0].copy()
    dEta = X[:,:,1].copy()
    dPhi = X[:,:,2].copy()
    dR = np.sqrt(dEta**2 + dPhi**2)
    
    z_collection[samp] = z.copy()
    eta_collection[samp] = dEta.copy()
    phi_collection[samp] = dPhi.copy()
    dR_collection[samp] = dR.copy()
    label_collection[samp] = Y[:,1].copy()
    
    del X, Y, z, dEta, dPhi, dR
```

```python
nbins = 15
avg_qjet_collection = {}
avg_gjet_collection = {}

for samp in samples:
    labels = label_collection[samp]
    dEta = eta_collection[samp]
    dPhi = phi_collection[samp]
    z = z_collection[samp]
    nq = np.count_nonzero(labels)
    ng = labels.size - nq
    avg_qjet = np.zeros((nbins,nbins))
    avg_gjet = np.zeros((nbins,nbins))
    eta_range = np.linspace(-0.2,0.2,num=nbins+1)
    phi_range = np.linspace(-0.2,0.2,num=nbins+1)
    for i, lab in enumerate(tqdm(labels)):
        jet_image = np.zeros((nbins,nbins))
        jtype = labels[i]
        etas = dEta[i]
        phis = dPhi[i]
        zs = z[i]
        for j in range(len(etas)):
            p_eta = etas[j]
            p_phi = phis[j]
            p_z = zs[j]
            eta_ind = -1
            phi_ind = -1
            for k in range(1,len(eta_range)):
                if p_eta >= eta_range[k-1] and p_eta < eta_range[k]:
                    eta_ind = k-1
                if p_phi >= phi_range[k-1] and p_phi < phi_range[k]:
                    phi_ind = k-1
            if eta_ind != 1 and phi_ind != -1:
                jet_image[eta_ind,phi_ind] += p_z
        if jtype == 1:
            avg_qjet += jet_image
        else:
            avg_gjet += jet_image
    avg_qjet /= nq
    avg_gjet /= ng
    
    avg_qjet_collection[samp] = avg_qjet
    avg_gjet_collection[samp] = avg_gjet
```

```python
plt.figure(1,(10,20))
for i, samp in enumerate(samples):
    avg_qjet = avg_qjet_collection[samp].copy()
    avg_gjet = avg_gjet_collection[samp].copy()
    plt.subplot(4,2,2*i+1)
    qimg = plt.imshow(avg_qjet.transpose(),origin='lower',vmin=0,vmax=0.12,extent=(-0.2,0.2,-0.2,0.2))
    plt.xlabel(r"$\Delta\eta$",fontsize=12)
    plt.ylabel(r"$\Delta\phi$",fontsize=12)
    plt.title("Avg Quark Jet, "+samp)
    plt.colorbar(qimg)
    plt.subplot(4,2,2*i+2)
    gimg = plt.imshow(avg_gjet.transpose(),origin='lower',vmin=0,vmax=0.12,extent=(-0.2,0.2,-0.2,0.2))
    plt.xlabel(r"$\Delta\eta$",fontsize=12)
    plt.ylabel(r"$\Delta\phi$",fontsize=12)
    plt.title("Avg Gluon Jet, "+samp)
    plt.colorbar(gimg)
plt.tight_layout()
```
