```python
exponents = np.linspace(0,15,num=100)
aucs = []
for exp in tqdm(exponents):
    opt = np.sum(np.power(z,exp,where=z>0,out=np.zeros(z.shape)),axis=1)
    auc = roc_auc_score(Y[:,1],opt)
    if auc < 0.5:
        auc = 1-auc
    aucs.append(auc)
```
