```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def parse_jets(file):
    jet_flav = []
    jet_z = []
    jet_theta = []
    jet_phi = []
    jet_death = []
    with open(file) as f:
        flav = []
        z = []
        theta = []
        phi = []
        death = []
        for l in f:
            if "BEGIN" in l:
                flav = []
                z = []
                theta = []
                phi = []
                death = []
            elif "END" in l:
                jet_flav.append(np.array(flav))
                jet_z.append(np.array(z))
                jet_theta.append(np.array(theta))
                jet_phi.append(np.array(phi))
                jet_death.append(np.array(death))
            else:
                vals = l.split(" ")
                flav.append(float(vals[0]))
                z.append(float(vals[1]))
                theta.append(float(vals[2]))
                phi.append(float(vals[3]))
                death.append(float(vals[4]))
    return jet_flav, jet_z, jet_theta, jet_phi
```

```python
base = '/global/homes/s/sambt/Jets/optimal-classifiers/duff_PS/no_gqq/Test/'
fq = base+'test_g_1overZ_expZ2test_largerZmin.txt'
fg = base+'test_g_1overZ_largerZmin.txt'
qf, qz, qth, qph = parse_jets(fq)
gf, gz, gth, gph = parse_jets(fg)
mq = [k.size for k in qf]
mg = [k.size for k in gf]
a,bins,patch = plt.hist(mq,histtype='step')
b = plt.hist(mg,bins=bins,histtype='step')
```

```python
obs_q = [np.sum(z*np.log(z)) for z in qz]
obs_g = [np.sum(z*np.log(z)) for z in gz]
a,bins,pat = plt.hist(obs_q,histtype='step',bins=100)
b = plt.hist(obs_g,bins=bins,histtype='step')
```
